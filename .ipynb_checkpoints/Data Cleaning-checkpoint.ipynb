### Droping columns having more than 50% null values

```python
df.drop(columns=['Verification' ,'WifiConnectivity'], inplace=True)
```

### Droping null rows from Content columns as it is our main foucs

```python
df.dropna(subset=['Content'], inplace=True)
```

### Replacing null values for numerical columns with mean and categorical columns with mode 

```python
def replace_null(col):
    if col.dtype == 'O':
        return col.fillna(col.mode()[0])
    else:
        return col.fillna(int(col.mean()))
```

```python
df[['Type_of_traveller','Route','DateFlown','SeatComfort','CabinStaffService','GroundService','FoodBeverages','InflightEntertainment']] = df[['Type_of_traveller','Route','DateFlown','SeatComfort','CabinStaffService','GroundService','FoodBeverages','InflightEntertainment']].apply(replace_null, axis= 0)
```

```python
df['Aircraft'] = df['Aircraft'].fillna('Unknown')
```

```python
df.isna().sum().apply(lambda x : str(round(x/df.shape[0]*100, 2))+" %")
```

```python
df.head(5)
```

```python
df['RatingValue'] = df['RatingValue'].apply(lambda x: int(x))
df['SeatComfort'] = df['SeatComfort'].apply(lambda x: int(x))
df['CabinStaffService'] = df['CabinStaffService'].apply(lambda x: int(x))
df['GroundService'] = df['GroundService'].apply(lambda x: int(x))
df['FoodBeverages'] = df['FoodBeverages'].apply(lambda x: int(x))
df['InflightEntertainment'] = df['InflightEntertainment'].apply(lambda x: int(x))
```

```python
df.info()
```

```python
df.head()
```

```python
df['Heading'] = df['Heading'].apply(lambda x: x.replace('"', ""))
df['Heading'] = df['Heading'].apply(lambda x: x.replace('“', ""))
df['Heading'] = df['Heading'].apply(lambda x: x.replace('”', ""))
```

```python
df
```

```python
df['Author'] = df['Author'].str.replace('\n', "")
```

```python
s = "Heathrow to San Francisco	"
ss = s.split("to")
ss = [x.strip() for x in ss]
ss
```

```python
def start_(val):
    vals = val.split("to")
    vals = [x.strip() for x in vals]
    return vals[0]
```

```python
df['Start'] = df['Route'].apply(start_)
```

```python
def end_(val):
    vals = val.split("to")
    vals = [x.strip() for x in vals]
    try:
        return vals[1]
    except:
        return None
```

```python
df['Destination'] = df['Route'].apply(end_)
```

```python
from dateutil import parser
```

```python
def date_fun(val):
    return parser.parse(val)
```

```python
df['Timeline'] = df['Timeline'].apply(date_fun)
```

```python
def date_month(val):
    val = parser.parse(val)
    val = val.strftime('%m/%y')
    return val
```

```python
df['DateFlown'] = df['DateFlown'].apply(date_month)
```
