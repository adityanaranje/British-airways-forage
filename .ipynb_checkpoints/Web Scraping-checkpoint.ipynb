## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup as bs
import requests
```

## Creat function to scrap data from url and store in lists

```python
ratingValues = []
headings = []
authors = []
countries = []
timelines = []
verifications = []
contentss = []
aircraft = []
type_of_travellers = []
seat_type = []
route = []
date_flown = []
seat_comfort = []
cabin_staff_service	= []
ground_service = []
value_for_money = []
recommended = []
food_and_beverages = []
inflight_entertainment = []
wifi_connectivity = []



def fetch_article(page_no):
    url = f"https://www.airlinequality.com/airline-reviews/british-airways/page/{page_no}/"
    response = requests.get(url)
    html_content = response.content

    soup = bs(html_content, 'lxml')
    articles = soup.find_all(itemprop = 'review')

    ## feathing all reviews on current page
    for article in articles:
        try:
            ratingValue =  int(article.find(itemprop = 'ratingValue').text) ## rating value    
            ratingValues.append(ratingValue)
        except:
            ratingValues.append(None)

        try:
            heading = str(article.find('h2', class_ = 'text_header').text)   ## heading
            headings.append(heading)
        except:
            headings.append(None)

        h3 = article.find('h3').text
        values = h3.split('(')
        
        try:
            author = values[0]   ## author
            authors.append(author)
        except:
            authors.append(None)

        try:
            country = values[1].split(')')[0] ## country
            countries.append(country)
        except:
            countries.append(None)
            
        try:
            timeline = values[1].split(')')[1]
            timelines.append(timeline)
        except:
            timelines.append(None)
        

        try:
            verification = article.find(class_ = 'text_content').find('em').text
            verifications.append(verification)
        except:
            verifications.append(None)

        try:
            contents = article.find(class_ = 'text_content').text
            content = contents.split('|')[1]
            contentss.append(content)
        except:
            contentss.append(None)
        
        table = article.find('table', class_ = 'review-ratings')
    
    
        try:
            val = table.find('td', class_ = 'review-rating-header aircraft').find_next_sibling().text
            aircraft.append(val)
        except:
            aircraft.append(None)

        try:
            val = table.find('td', class_ = 'review-rating-header type_of_traveller').find_next_sibling().text
            type_of_travellers.append(val)
        except:
            type_of_travellers.append(None)

        try:
            val = table.find('td', class_ = 'review-rating-header cabin_flown').find_next_sibling().text
            seat_type.append(val)
        except:
            seat_type.append(None)
        
        try:
            val = table.find('td', class_ = 'review-rating-header route').find_next_sibling().text
            route.append(val)
        except:
            route.append(None)
        
        try:  
            val = table.find('td', class_ = 'review-rating-header date_flown').find_next_sibling().text
            date_flown.append(val)
        except:
            date_flown.append(None)


        try:
            val = table.find('td', class_ = 'review-rating-header recommended').find_next_sibling().text
            recommended.append(val)
        except:
            recommended.append(None)

        
        try:    
            val = table.find('td', class_ = 'review-rating-header seat_comfort').find_next_sibling().find(class_ = 'fill').text
            seat_comfort.append(int(val))
        except:
            seat_comfort.append(None)
        
        try:
            val = table.find('td', class_ = 'review-rating-header cabin_staff_service').find_next_sibling().find(class_ = 'fill').text
            cabin_staff_service.append(int(val))
        except:
            cabin_staff_service.append(None)
        
        try:
            val = table.find('td', class_ = 'review-rating-header ground_service').find_next_sibling().find(class_ = 'fill').text
            ground_service.append(int(val))
        except:
            ground_service.append(None)
        
        try:
            val = table.find('td', class_ = 'review-rating-header value_for_money').find_next_sibling().find(class_ = 'fill').text
            value_for_money.append(int(val))
        except:
            value_for_money.append(None)
        

        try:
            val = table.find('td', class_ = 'review-rating-header inflight_entertainment').find_next_sibling().find(class_ = 'fill').text
            inflight_entertainment.append(int(val))
        except:
            inflight_entertainment.append(None)


        try:
            val = table.find('td', class_ = 'review-rating-header wifi_and_connectivity').find_next_sibling().find(class_ = 'fill').text
            wifi_connectivity.append(int(val))
        except:
            wifi_connectivity.append(None)


        try:
            val = table.find('td', class_ = 'review-rating-header food_and_beverages').find_next_sibling().find(class_ = 'fill').text
            food_and_beverages.append(int(val))
        except:
            food_and_beverages.append(None)

        
```

## Calling function to fetch data all pages using loop 

```python
%%time
for i in range(1,382): 
    fetch_article(i)
```

## Creating dataframe to store all lists in dataframe

```python
df = pd.DataFrame()
```

```python
df['Heading'] = headings
df['RatingValue'] = ratingValues
df['Author'] = authors
df['Country'] = countries
df['Timeline'] = timelines
df['Verification'] = verifications
df['Content'] = contentss
df['Aircraft'] = aircraft
df['Type_of_traveller'] = type_of_travellers
df['SeatType'] = seat_type
df['Route'] = route
df['DateFlown'] = date_flown
df['SeatComfort'] = seat_comfort
df['CabinStaffService'] = cabin_staff_service
df['GroundService'] = ground_service
df['ValueForMoney'] = value_for_money
df['Recommend'] = recommended
df['FoodBeverages'] = food_and_beverages
df['InflightEntertainment'] = inflight_entertainment
df['WifiConnectivity'] = wifi_connectivity
```

## Saving dataframe as CSV file

```python
df.to_csv('British_Airways.csv', header=True, index=False)
```
